```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#converting a csv file into a pandas dataframe
```

```python
diabetes_data=pd.read_csv(r'C:\Users\jerem\Desktop\ml data\diabetes.csv')
diabetes_data
```

```python
#checking how many people are diabetic and how many are none diabetic
```

```python
diabetes_data['Outcome'].value_counts()
```

```python
#checking for null values in our dataset
```

```python
diabetes_data.isnull().sum()
```

```python
diabetes_data.values.std()
```

```python
diabetes_data.groupby('Outcome').mean()
```

```python
#seperating the data from the labels
```

```python
x=diabetes_data.drop(columns='Outcome',axis=1)
y=diabetes_data['Outcome']
print (x)
print (y)
```

```python
#standardizing the data
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
print(x.std())
```

```python
#splitting the data into training data and testing data
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.1,stratify=y,random_state=2)
print(x.shape)
print(x_train.shape)
print(y_test.shape)
```

```python
#training our model
```

```python
classifier=svm.SVC(kernel='linear')
classifier.fit(x_train,y_train)
```

```python
#evaluating our model
```

```python
train_prediction=classifier.predict(x_train)
train_accuracy=accuracy_score(train_prediction,y_train)
print(train_accuracy)
test_prediction=classifier.predict(x_test)
test_accuracy=accuracy_score(test_prediction,y_test)
print(test_accuracy)
```

```python
#creating a diabetes predicting system
```

```python
input_data=(3,187,70,22,200,36.4,0.408,36)
array_data=np.asarray(input_data)
reshaped_data=array_data.reshape(1,-1)
```

```python
std_input=scaler.transform(reshaped_data)

prediction=classifier.predict(std_input)
if (prediction[0]==1):
    print ('The patient is Diabetic')
elif (prediction[0]==0):
    print('the patient is non-diabetic')
```
